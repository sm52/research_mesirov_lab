```python
import pandas as pd
import matplotlib.pyplot as mat

def data_read(file, summ = False):
    data = pd.read_csv(file, sep='\t',skiprows=2)
    if(summ == True):
        print("Rows " + str(len(data.index)))
        print("Columns " + str(len(data.columns)))
    return data
    
minimal = data_read("BRCA_minimal_60x19.gct", True)

large = data_read("BRCA_large_20783x40.gct", True)

large.columns

fig, ((ax0, ax1), (ax2, ax3)) = mat.subplots(nrows=2, ncols=2)

ax0.hist(large["A7-A0DB-normal"])
ax1.hist(large["A7-A13E-normal"])
ax2.hist(large["BH-A0B3-primary"])
ax3.hist(large["BH-A0B5-primary"])
    
```
